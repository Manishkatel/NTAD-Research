```python
import pandas as pd
from pycaret.classification import *

# Load your preprocessed dataset
df = pd.read_csv(r"C:\Users\manis\Downloads\preprocessed_multiple_class.csv")

# Target column
target = 'attack type'

# Show quick overview
print(df.shape)
df.head() 
```
